RGB --> GrayScale

```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('/content/steph-curry_Slice_RLV_2022-1600x1066.jpg')

# Getting the R,G and B dimension matrices
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
# RGB to grayscale conversion formula 
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(imgGray, cmap='gray')
plt.show()




```

RGB --> hsv

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
  
  
image = mpimg.imread('/content/steph-curry_Slice_RLV_2022-1600x1066.jpg')

  
  
hsv = matplotlib.colors.rgb_to_hsv(image)
plt.imshow(hsv)
```

rgb --> sketch

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')

# Loading Original Image
img = cv2.imread('/content/steph-curry_Slice_RLV_2022-1600x1066.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


#Converting Image to GrayScale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


#Inverting the Image
img_invert = cv2.bitwise_not(img_gray)


#Converting the image to sketch
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)



final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.show()
```
